```r
data=read.csv('../input/water_dataX.csv',check.names=F)
```

```r
colnames(data)[5] = "DO"
colnames(data)[7] = "CONDUCTIVITY"
colnames(data)[8] = "BOD"
colnames(data)[9] = "NITRATE_NITRITE"
colnames(data)[10] = "FECAL_COLIFORM"
colnames(data)[11] = "TOTAL_COLIFORM"
```

```r
df=data
dim(df)
```

```r
df
```

```r
df$DO=as.numeric(df$DO)
df$PH=as.numeric(df$PH)
df$CONDUCTIVITY=as.numeric(df$CONDUCTIVITY)
df$BOD=as.numeric(df$BOD)
df$NITRATE_NITRITE=as.numeric(df$NITRATE_NITRITE)
df$TOTAL_COLIFORM=as.numeric(df$TOTAL_COLIFORM)
```

```r
library(dplyr)
```

```r
df$nPH=case_when(
  6.7>=df$PH & df$PH>6.5 ~ 40,
  6.8>=df$PH & df$PH>6.7 ~ 60,
  6.9>=df$PH & df$PH>6.8 ~ 80,
  8.5>=df$PH & df$PH>=7 ~ 100,
  8.6>=df$PH & df$PH>8.5 ~ 80,
  8.8>=df$PH & df$PH>8.6 ~ 60,
  9>=df$PH & df$PH>8.8 ~ 40,
  TRUE ~ 0
)
```

```r
df$nDO=case_when(
    df$DO>=6 ~ 100,
    6>=df$DO & df$DO>=5.1 ~ 80,
    5.1>df$DO & df$DO>=4.1 ~ 60,
    4.1>df$DO & df$DO>=3 ~ 40,
    TRUE ~ 0
)
```

```r
df$nBOD=case_when(
    3>=df$BOD & df$BOD>0 ~ 100,
    6>=df$BOD & df$BOD>3 ~ 80,
    80>=df$BOD & df$BOD>6 ~ 60,
    125>=df$BOD & df$BOD>80 ~ 40,
    TRUE ~ 0
)
```

```r
df$nEC=case_when(
    75>=df$CONDUCTIVITY & df$CONDUCTIVITY>0 ~ 100,
    150>=df$CONDUCTIVITY & df$CONDUCTIVITY>75 ~ 80,
    225>=df$CONDUCTIVITY & df$CONDUCTIVITY>150 ~ 60,
    300>=df$CONDUCTIVITY & df$CONDUCTIVITY>225 ~ 40,
    TRUE ~ 0
)
```

```r
df$nNN=case_when(
    20>=df$NITRATE_NITRITE & df$NITRATE_NITRITE>0 ~ 100,
    50>=df$NITRATE_NITRITE & df$NITRATE_NITRITE>20 ~ 80,
    100>=df$NITRATE_NITRITE & df$NITRATE_NITRITE>50 ~ 60,
    200>=df$NITRATE_NITRITE & df$NITRATE_NITRITE>100 ~ 40,
    TRUE ~ 0
)
```

```r
df$nTC=case_when(
    5>=df$TOTAL_COLIFORM & df$TOTAL_COLIFORM>0 ~ 100,
    50>=df$TOTAL_COLIFORM & df$TOTAL_COLIFORM>5 ~ 80,
    500>=df$TOTAL_COLIFORM & df$TOTAL_COLIFORM>50 ~ 60,
    1000>=df$TOTAL_COLIFORM & df$TOTAL_COLIFORM>500 ~ 40,
    TRUE ~ 0
)
```

```r
df$wPH=df$nPH*0.16
df$wDO=df$nDO*0.28
df$wTC=df$nTC*0.28
df$wBOD=df$nBOD*0.23
df$wEC=df$nEC*0.09
df$wNN=df$nNN*0.028
df$wQI=df$wPH+df$wDO+df$wTC+df$wBOD+df$wEC+df$wNN
```

```r
hist(df$wQI, col = "red")
```

```r
hist(df$year, col = "red")
```

```r
agg=df %>%
  group_by(year)%>%
  summarise(mean(wQI))
```

```r
library(ggplot2)
```

```r
ggplot(df,aes(x=factor(df$year),y=df$wQI)) + geom_boxplot()
```

```r
ggplot(agg,aes(x=factor(agg$year),y=agg$`mean(wQI)`)) + geom_point()
```

```r
targetdata = agg[,c(1,2)]
a = sample(nrow(targetdata),nrow(targetdata)*0.7)
train = targetdata[a,]
test = targetdata[-a,]
```

```r
model1 = lm(agg$`mean(wQI)` ~ agg$year,data = train)
summary(model1)
ypred = predict(model1,type="response", test)
SSE = sum((agg$`mean(wQI)` - ypred) ^ 2)
SST = sum((agg$`mean(wQI)` - mean(agg$`mean(wQI)`)) ^ 2)
Rsquare <-( 1 - (SSE/SST))
print(Rsquare)
```
